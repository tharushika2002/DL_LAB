### 1. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.

```python
excepcol = ['N_total', 'N_bulk' , 'N_surface' , 'R_avg']
```

```python
new_nano_data = nano_data.drop(columns = excepcol)
```

```python
new_nano_data.shape
```

### 2. Display the first 20 samples of this dataframe

```python
new_nano_data.sample(20)
```

### 3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.

```python
mean = nano_data[excepcol].mean()
```

```python
print (f'All means \n {mean}')
```

```python
std = nano_data[excepcol].std()
```

```python
print (f'all std \n {std}')
```

```python
allQuartile = nano_data[excepcol].quantile([0.25,0.50,0.75])
print (allQuartile)
```

```python
import matplotlib 
import matplotlib.pyplot as plt
```

```python
fig , axes = plt.subplots(nrows=1, ncols=4 , figsize=(20,4))

for i in range(4):
    axes[i].hist(nano_data[excepcol[i]], bins=30, color='Blue', edgecolor = 'Yellow')
    axes[i].set_title(f'Histogram {i+1}')
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel('Frequency')
```
